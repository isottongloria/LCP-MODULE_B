### Plot the colorbar for the best weight

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# we should find the best model with the highest accurecy
#and get the best accurecy "w" 
w_best=np.loadtxt("weight -1") 
```

```python
amn=5  #the amino
s_amn=4  # sub layer of amino

M=2 #didden units

o1,o2,o3,o4= np.eye(s_amn)
```

```python
my_w= w_best.reshape((amn,s_amn,M))
```

```python
out= np.zeros((s_amn,amn,M))

for i in range(M):
    out[:,:,i]=np.array([
        np.dot(my_w[:,:,i],o1),
        np.dot(my_w[:,:,i],o2),
        np.dot(my_w[:,:,i],o3),
        np.dot(my_w[:,:,i],o4)
        ])
    #print(out)
    
```

```python
fig, ax= plt.subplots(nrows=1, ncols=2,figsize=(8,4),dpi=120)
#fig.tight_layout()

for i in range (M):
    m=ax[i].matshow(out[:,:,i],cmap='plasma')
    ax[i].set_ylabel("Type of Amino-Acid")


cbar=fig.add_axes([0.98,0.15,0.05, 0.7])
fig.colorbar(m, cbar)
plt.show()
```
